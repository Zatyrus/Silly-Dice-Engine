```python
## Check uniformity
max: int = 6
sample:int = 1000
sample:list = [draw_uniform(max = max,
                            generator= gen) 
               for _ in range(sample)]

fig, ax = plt.subplots(1,1, figsize = (8,4), dpi = 100)
ax.hist(np.array(sample),
        bins = 6)

ax.legend(['test'])

plt.show()
print(sps.expectile(sample))
```

```python
## Check explode
max: int = 6
sample:int = 10000
sample:list = [explode_uniform(max = max,
                            generator= gen) 
               for _ in range(sample)]

fig, ax = plt.subplots(1,1, figsize = (8,4), dpi = 100)
ax.hist(np.array(sample),
        bins = 25)

plt.show()
print(sps.expectile(sample))
```
